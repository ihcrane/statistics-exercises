```python
from env import username, hostname, password

def get_db_url(username, hostname, password, database):
    url = f'mysql+pymysql://{username}:{password}@{hostname}/{database}'
    return url
```

```python
url = get_db_url(username, hostname, password, 'employees')
```

```python
salaries = pd.read_sql('SELECT * FROM salaries WHERE to_date >= CURDATE()', url)
salaries.head()
```

```python
salaries = pd.DataFrame(salaries)
salaries
```

```python
mean = salaries['salary'].mean()
mean
```

```python
sd = salaries['salary'].std()
sd
```

```python
salaries['salary'].median()
```

```python
sns.histplot(salaries['salary'], kde=True)
```

```python
stats.norm(mean, sd).cdf(60000)
```

```python
stats.norm(mean, sd).sf(95000)
```

```python
(stats.norm(mean, sd).cdf(80000)) - (stats.norm(mean, sd).cdf(65000))
```

```python
stats.norm(mean, sd).isf(.05)
```

```python
stats.norm(salaries['salary']).sf(95000).mean()
```

```python
(stats.norm(salaries['salary']).cdf(80000).mean()) - (stats.norm(salaries['salary']).cdf(65000).mean())
```
